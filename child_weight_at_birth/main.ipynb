```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
ds = pd.read_csv("data.csv")
ds
```

```python
# clear all na
ds['smoke'].fillna(3,inplace=True)
ds['weight'].fillna(0,inplace=True)
ds['height'].fillna(0,inplace=True)
ds['gestation'].fillna(0,inplace=True)
ds
```

```python
sz = math.ceil(ds[ds.columns[0]].count()*0.7)
training_set = ds.sample(n=sz)
test_set  = []
for i in range(ds[ds.columns[0]].count()):
    try:
        training_set.loc[i]
    except:
        test_set.append(list(ds.loc[i]))
random.shuffle(test_set)
mid = int(len(test_set)/2)
validation_set = test_set[mid:]
test_set = test_set[:mid]
```
